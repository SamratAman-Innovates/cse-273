# Groupby in Pandas

```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Name": ["Aman", "Saurav", "Karan", "Shubham","Abhishek",None, "Sameer"],
    "Age": [22, 23, None, 22, 25, 24, 26],
   "Marks": [85, None, 88, 90, 56, 72, None],
   "Subject": ["Math", "Physics", "Maths", "Chemistry", "Math", "Physics", "Chemistry"]
}
df = pd.DataFrame(data)
df
```

##  basic group by


```python
df.groupby("Subject")["Marks"].mean()   # mean marks per subject and it automatically ignor nan So missing marks will not break calculation.
```

```python
df.groupby("Subject")["Name"].count() # count student per subject ignoring missing value by default
```

```python
df.groupby("Subject").size()  # this count with missing value also
```

```python
# multiple agrregration with single column
df.groupby("Subject")["Marks"].agg(["mean","max","min","sum","count"])
```

```python
#multiple agrregration by multiple column
df.groupby(["Subject","Age"])["Marks"].mean()
```
