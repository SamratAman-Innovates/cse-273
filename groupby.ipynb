## Pandas GroupBy


```python
import numpy as np
import pandas as pd
```

```python
#you are hr data analyst for techCrop . you have been provided with a dataframe as df 
# name employ name
#department 
# role 
# experimence
#salary 
#score
data = {
    "Name": ["Aman","Sagar","Sehab","Samrat"],
    "Department": ["IT","HR","Finance","Marketing"],
    "Role": ["Developer","Manager","Analyst","Specialist"],
    "Experimence": [10,5,8,12],
    "Salary": [50000,60000,55000,70000],
    "Score": [85,92,78,95]
}
df = pd.DataFrame(data)
df
```

```python
#basic average 
#mean salary for each department
df.groupby("Department")["Salary"].mean()
```

```python
# mean performance score for each role
df.groupby("Role")["Score"].mean()
```

```python
#counting and sizing 
# how many actual non null performenace score entries exit for each deparment
df.groupby("Department")["Score"].count()
```

```python
# what is the total number of employee record(row) in each department, regardless of whether data is missing
df.groupby("Department").size()
```

```python
#multiple aggregate
#geenrate table foe salary based on department thta show mean , maximu, minumu and count in asingle line of code
df.groupby("Department")["Salary"].agg(["mean", "max", "min", "count"])
```

```python
# generate the same summummry table (mean,max,min,count) but analyze the performance Score based on the role
df.groupby("Role")["Score"].agg(["mean", "max", "min", "count"])
```

```python
#grouping by multiple columns
#find the average salary grouped by both department and role.
df.groupby(["Department","Role"])["Salary"].mean()
```

```python
#find average performance score grouped by both department and experience
df.groupby(["Department","Experimence"])["Score"].mean()
```
