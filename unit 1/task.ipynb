```python
# acces in numpy

arr = np.array([[1,2,3],
               [4,5,6]])
print(arr)
print(np.sum(arr, axis=0))
# print(sum(arr))
```

```python
#where is numpy
result = np.where(arr < 0, 0 , arr)  #for arr <0 conditon , 0 for true , arr for false
print(result)
```
