```python
#S1 #S2 #S3
import numpy as np
marks=np.array(([78,85,90],#Student 1
                [45,60,55],#Student 2
                [88,92,95],#Student 3
                [30,40,35],#Student 4
                [65,70,75]))#Student 5
marks
```

```python
print(marks.shape)
print(marks.size)
print(marks.dtype)
```

```python
#exact all marks of 2nd student
second = marks[1:2,:]
print(second)
```

```python
three = marks[0:,2:]
print(three)
```

```python
first = marks[0:1,1:2]
print(first)
```

```python
#calculate the avg marks of each student
avg_marks = np.mean(marks, axis=1)
print(avg_marks)
```

```python
#calculate the avg marks of each subject
avg_per_subject = np.mean(marks, axis=0)
print(avg_per_subject)
```

```python
#task c1 pass / fails mapping
pass_fail = np.where(marks >= 50, 'Pass', 'Fail')
```

```python
pass_s = avg_marks >= 50
print(pass_fail)
```

```python
fail_s = avg_marks < 50
print(fail_s)
```

```python
#grade allocation
#use nested condition logic within numpy
grades = np.where(avg_marks >= 80, 'A',
                  np.where(avg_marks >= 70, 'B',
                           np.where(avg_marks >= 50, 'C',
                                    np.where(avg_marks < 50, 'D', 'F'))))

grades
```

```python
grades = np.where(marks < 50, 0, np.where(marks < 70, 6, np.where(marks < 85 , 8, 10)))
grades
```
