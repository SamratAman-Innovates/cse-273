```python
#1 technical q - look at the marks colunmn in your output why are the values floats(eg , 85.0) instead of integers?
# ans---->> This happens because pandas uses float64 dtype to represent missing values (NaN) in numeric columns. 
# When there are missing values, pandas must use a floating-point type to accommodate both integers and NaN values in the same column.
```

```python
# 2 technical q  - what is the risk of using inplace = true when droping missing values?
# Risk: Using inplace=True modifies the original DataFrame directly,
#  which can lead to loss of data if the operation is not carefully planned. 
# It also makes debugging more difficult because the original data is altered in place, 
# and there's no easy way to revert changes without re-reading the original data.
```
